### Normalizo el dataset

- Comienzo con las mayusculas

```python
df_mapa.columns = df_mapa.columns.str.lower()
```

- Eliminar Espacios en Blanco y Reemplazar con Guiones Bajos

```python
df_mapa.columns = df_mapa.columns.str.replace(' ', '_')
```

- Elimino caracteres especiales

```python
df_mapa.columns = df_mapa.columns.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df_mapa.columns = df_mapa.columns.str.replace(r'[^\w\s]', '', regex=True)
```

- Aplico el mismo proceso en las columnas 'provincia', 'partido', 'localidad' y 'poblacion'

```python
# Normalizar valores en columnas específicas
df_mapa['provincia'] = df_mapa['provincia'].str.lower().str.replace(' ', '_').str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8').str.replace(r'[^\w\s]', '', regex=True)
df_mapa['partido'] = df_mapa['partido'].str.lower().str.replace(' ', '_').str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8').str.replace(r'[^\w\s]', '', regex=True)
df_mapa['localidad'] = df_mapa['localidad'].str.lower().str.replace(' ', '_').str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8').str.replace(r'[^\w\s]', '', regex=True)
```

```python
df_mapa
```

Reemplazo los guiones '--' por 'NO'

```python
# Reemplazar '--' por 'NO'
df_mapa.replace('--', 'NO', inplace=True)
df_mapa
```

- Convierto la columna 'link' a entero

```python
df_mapa['link'] = df_mapa['link'].astype(int)
```

- Renombro la columna 'link' a 'link_indec' para unirlo con notebook 'etleda_hoja2'

```python
df_mapa.rename(columns={'link': 'link_indec'}, inplace=True)
```
